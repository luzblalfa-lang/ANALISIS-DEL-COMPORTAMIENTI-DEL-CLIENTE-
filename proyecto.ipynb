```python
#identificar los nulo
df.isnull().sum()
```

```python
mediana=df['Review Rating'].median()
print(mediana)
```

```python
#imputar valores nulos
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))
```

```python
#identificar los nulo
df.isnull().sum()
```

```python
#cambiamos a minusculas y llenamos los espacios en blanco con guion abajo a los titulos o campos
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
import pandas as pd
bins = [0,17,34,60,100]
labels = ['menor de edad','joven','adulto','senior']

#etiquetado
df['grupo_edad']=pd.cut(df['age'],bins=bins,labels=labels)
```

```python
#mostrar las columnas
df[['age','grupo_edad']].head(10)
```

```python
df
```

```python
#creamos nueva columna para clasificar las frecuencias
frecuency_mapping ={
    'Fortnightly':14,
    'Weekly':7,
    'Annually':365,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Monthly':30,
    'Every 3 Months':90
}

#crear columna metodo map
df['frequency_of_purchases:days']=df['frequency_of_purchases'].map(frecuency_mapping)
```

```python
df[['frequency_of_purchases:days','frequency_of_purchases']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
#eliminamos una de ellas
df=df.drop('promo_code_used',axis=1)
```

```python
df.columns
```

```python
df
```

```python
#cambiar el nombre de las columnas

diccionario = {
    "customer_id":"idcliente",
    "age":"edad",
    "gender":"sexo",
    "item_purchased":"articulo_comprado",
    "category":"categoria",
    "purchase_amount":"monto_de_compra",
    "location":"ubicacion",
    "size":"tamaño",
    "color":"color",
    "season":"temporada",
    "review_rating":"calificacion_del_cliente",
    "subscription_status":"estado_de_suscripcion",
    "shipping_type":"tipo_de_envio",
    "discount_applied":"descuento_aplicado",
    "previous_purchases":"compras_anteriores",
    "payment_method":"metodo_de_pago",
    "frequency_of_purchases":"frecuencia_de_compra",
    "grupo_edad":"grupo_edad",
    "frequency_of_purchases:days":"frecuencia_de_compras_dias"

}

df=df.rename(columns=diccionario)
```
